```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Animal Dataset.csv")
```

```python
df
```

```python
df['Height (cm)']= df['Height (cm)'].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
```

```python
df
```

```python
df["Weight (kg)"]=df["Weight (kg)"].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
df["Lifespan (years)"]=df["Lifespan (years)"].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
df["Average Speed (km/h)"]=df["Average Speed (km/h)"].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
df["Gestation Period (days)"]=df["Gestation Period (days)"].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
df["Top Speed (km/h)"]=df["Top Speed (km/h)"].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
df["Offspring per Birth"]=df["Offspring per Birth"].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axis=1)
```
