```python
df['Height (kg)']= df['Height (kg)'].str.extract(r'(\d+)-?(\d+)?').astype(float).mean(axix=1)
```

```python
df['Weight (kg)'] = pd.to_numeric(df['Weight (kg)'], errors='coerce').astype(float)
```

```python
df.info()
```

```python
df['Lifespan (years)'] = pd.to_numeric(df['Lifespan (years)'], errors='coerce').astype(float)
df['Average Speed (km/h)'] = pd.to_numeric(df['Average Speed (km/h)'], errors='coerce').astype(float)
df['Gestation Period (days)'] = pd.to_numeric(df['Gestation Period (days)'], errors='coerce').astype(float)
df['Top Speed (km/h)'] = pd.to_numeric(df['Top Speed (km/h)'], errors='coerce').astype(float)
df['Offspring per Birth'] = pd.to_numeric(df['Offspring per Birth'], errors='coerce').astype(float)
```
